```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

# Question 1

```python
def stockdata_in_array():
    f = open('data.csv','r')
    data = []
    stockNames=[]
    nb_line = 0
    for line in f :
        u = [x for x in line.split(",")]
        data.append(u)
        stockNames.append(data[nb_line][0])
        del data[nb_line][0] #delete the name of each stock
        nb_line += 1 
    del data[0:1] #delete the first line with 'StockName' and the dates
    del stockNames[0:1] #delete the first line with 'StockName'
    dataArray = np.array(data, dtype='float32')
    return dataArray, data, stockNames

dataArray, data, stockNames = stockdata_in_array()

def kmeans_on_stockdata(k, max_i, n_i):
    kmeans = KMeans(init="k-means++", n_clusters=k, max_iter=max_i, n_init=n_i).fit(dataArray)
    return kmeans.labels_, kmeans.cluster_centers_
```

```python
def SSE_stockdata(k, max_i, n_i):
    clusters, centroids = kmeans_on_stockdata(k, max_i, n_i)
    SSE = 0
    for i in range(k):
        for j in range (len(dataArray)):
            if clusters[j] == i : #choose all the points of the the cluster number i
                for l in range (len(dataArray[i])):
                    SSE += ((dataArray[j][l]-centroids[i][l])**2)
    return SSE
```

```python
SSE_stockdata(8, 10000, 100)
```

# Question 2

We want to decrease the SSE : we are going to play with the values of the parameters max_iter and n_init !
First, we are going to change the parameter max_iter from 1 to 10000 while n_ini is fixed to 100. max_iter is the maximum number of iterations of the k-means algorithm for a single run.

```python
max_iter = []
SSE = []
for i in range(1,20001, 100):
    max_iter.append(i)
    SSE.append(SSE_stockdata(8, i, 100))
plt.plot(max_iter, SSE)
plt.xlabel("max_iter")
plt.ylabel("SSE")
plt.show()
```

We can observe that the parameter max_iter has an uncertain impact on the SSE (2 close values can give a very different SSE). Increasing or decreasing the maximum of iterations of the algrithm for a  doesn't have a big impact on SSE. We are going to keep max_iter = 10000.

Now we are going to change the value of n_init from 1 to 10000 while max_iter is fixed to 10000. n_init is the number of time the k-means algorithm will be run with different centroid seeds.

```python
n_init = []     
SSE_n = []
for i in range(1,1000, 50):
    n_init.append(i)
    SSE_n.append(SSE_stockdata(8, 10000, i))
plt.plot(n_init, SSE_n)
plt.xlabel("n_init")
plt.ylabel("SSE")
plt.show()
```

We can observe that the SSE tends to decrease as n_init increases for the small values of n_init and then we reach a level. However, givent that the bigger n_init is, the more the algorithm will take time to run, we are going to take n_init = 400.
